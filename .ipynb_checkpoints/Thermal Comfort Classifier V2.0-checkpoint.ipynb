## Data Acquisition

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/mzhafranaf/Final-Project-Orbit/main/darjeeling.csv"
dataset = pd.read_csv(url)
```

```python
dataset.head()
```

```python
dataset.info()
```

## Data Cleaning

```python
dataset.isnull().sum()
```

There's unnamed feature that should remove from our data

```python
dataset.drop(dataset.columns[dataset.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
dataset.info()
```

There's some numeric features that's declared as object type. So, we should convert it to float64 type.

```python
data = dataset.dropna()
#data
```

```python
# remove Location and Subject Code
data = data.drop(['Location','Subject Code'], axis=1)
```

```python
objectFeatures = ['TP','air_ts', 'air_tp', 'comfort', 'prod', 'TA', 'Uphols', 'sh_sw']

for i in objectFeatures:
    data[i] = pd.to_numeric(data[i],errors = 'coerce')
data.info()
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

Because we want to focused on TSV, we should know how many data for each class in TSV

```python
data.TSV.unique()
```

```python
data['TSV'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.countplot(x='TSV', data=data, order=data.TSV.unique().sort())
```

Because distribution data of each class is uneven, we should simplify it from 7 classes to 3 classes
* Class cold (0) : -3, -2, -1
* Class neutral (1) : 0
* Class warm (2) : 1, 2, 3

```python
comfortLevel = { 3.0:2,
                 2.0:2,
                 1.0:2,
                 0.0:1,
                -1.0 :0,
                -2.0:0,
                -3.0:0}

data['TSV'] = data['TSV'].map(comfortLevel)
data.head()
```

```python
sns.countplot(x='TSV', data=data, order=data.TSV.unique().sort())
```

# Classification

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.inspection import permutation_importance
```

## Data Splitting

```python
y = data[['TSV']]
X = data.drop(['TSV'], axis=1)

# scaler = StandardScaler()
# X = scaler.fit_transform(X)

scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X),columns = X.columns)
```

## Splitting Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

## Modeling

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)

importance = model.feature_importances_
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i,v))
    
plt.figure(figsize=(12,9))
plt.barh([x for x in range(len(importance))], importance)
plt.xlabel('Score')
plt.ylabel('Feature')
plt.show()
```

```python
result = permutation_importance(model, X_train, y_train, n_repeats=10, random_state=42)
perm_sorted_idx = result.importances_mean.argsort()

tree_importance_sorted_idx = np.argsort(model.feature_importances_)
tree_indices = np.arange(0, len(model.feature_importances_)) + 0.5

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 8))
ax1.barh(tree_indices, model.feature_importances_[tree_importance_sorted_idx], height=0.7)
ax1.set_yticks(tree_indices)
ax1.set_yticklabels(data.columns[tree_importance_sorted_idx])
ax1.set_ylim((0, len(model.feature_importances_)))
ax2.boxplot(
    result.importances[perm_sorted_idx].T,
    vert=False,
    labels=data.columns[perm_sorted_idx],
)
fig.tight_layout()
plt.show()
```
